# Image Cropping

```python
from matplotlib import pyplot as plt
import cv2
```

```python
img = plt.imread("circles.jpg")
```

```python
plt.imshow(img)
```

```python
plt.show()
```

# Image에 Cropping을 위해 일정한 여백을 만들어줄 수 있습니다.

```python
help(cv2.copyMakeBorder)
```

```python
bordersize = 100
color = [255, 255, 255]
bordered_img=cv2.copyMakeBorder(img, top=bordersize, bottom=bordersize, left=bordersize, right=bordersize, borderType= cv2.BORDER_CONSTANT, value=color )
```

```python
#cv2.circle(bordered_img, (300, 320), radius = 10, color = [255, 0 ,0])
plt.imshow(bordered_img)
plt.show()
```

# 때로 cropping을 하기 전 회전을 해주기도 합니다.

```python
center = (300, 320)
angle = 45
```

```python
rot_mat = cv2.getRotationMatrix2D(center, angle, 1.0) #cropped image size
result = cv2.warpAffine(bordered_img, rot_mat, (bordered_img.shape[1], bordered_img.shape[0]),flags=cv2.INTER_LINEAR)

plt.imshow(result)
plt.show()
```

# 일정한 patch size를 가지고 잘라줄 수 있습니다.

```python
patchsize = 400
patch = (int(patchsize), int(patchsize))
center = (300, 320)
cropped_image = cv2.getRectSubPix(result, patch, center)
plt.imshow(cropped_image)
plt.show()
```
