### Global Histogram Equalization 
It is a nonlinear operation used to improve the contrast of images. e.g., when the pixel values of images are confined to some specific range of values only.

An idea is like <img src="http://docs.opencv.org/trunk/_images/histogram_equalization.png" />

```python
## very simple and straight interface

wiki = cv2.imread("public-images/wiki.jpg", cv2.IMREAD_GRAYSCALE)
equalized_wiki = cv2.equalizeHist(wiki, )

fig, axes = plt.subplots(1, 2, figsize = (10, 5))
axes[0].imshow(wiki, cmap = plt.cm.gray)
axes[1].imshow(equalized_wiki, cmap = plt.cm.gray)
```

### Adaptive Histogram Equalization 

**CLAHE (Contrast Limited Adaptive Histogram Equalization)**

adaptive histogram equalization is used. In this, image is divided into small blocks called “tiles” (tileSize is 8x8 by default in OpenCV). Then each of these blocks are histogram equalized as usual. 

So in a small area, histogram would confine to a small region (unless there is noise). If noise is there, it will be amplified. To avoid this, contrast limiting is applied. If any histogram bin is above the specified contrast limit (by default 40 in OpenCV), those pixels are clipped and distributed uniformly to other bins before applying histogram equalization. After equalization, to remove artifacts in tile borders, bilinear interpolation is applied.

```python
clahe = cv2.imread("public-images/clahe.png", cv2.IMREAD_GRAYSCALE)

global_equalized = cv2.equalizeHist(clahe)

claher = cv2.createCLAHE(clipLimit = 2.0, tileGridSize = (8, 8))
adaptive_equalized = claher.apply(clahe)

fig, axes = plt.subplots(3, 1, figsize = (6, 6 * 3))
axes[0].imshow(clahe, cmap = plt.cm.gray)
axes[1].imshow(global_equalized, cmap = plt.cm.gray)
axes[2].imshow(adaptive_equalized, cmap = plt.cm.gray)
```

```python

def explore_clahe(clipLimit, tileGridSize):
    

    claher = cv2.createCLAHE(clipLimit = clipLimit, tileGridSize = (tileGridSize, tileGridSize))
    adaptive_equalized = claher.apply(clahe)

    fig, axes = plt.subplots(1, 2, figsize = (6*2, 6))
    axes[0].imshow(clahe, cmap = plt.cm.gray)
    axes[0].set_title("original")
    axes[1].imshow(adaptive_equalized, cmap = plt.cm.gray)
    axes[1].set_title("clahed")
    
from IPython.html import widgets
widgets.interact(explore_clahe, clipLimit=(0.1, 6.0, 0.5), tileGridSize = (1, 12, 1))
```

### Histogram backprojection (probability matrix/image)

It is used for image segmentation or finding objects of interest in an image (e.g., of uniform color patterns). In simple words, it creates an image of the same size (but single channel) as that of our input image, where each pixel corresponds to the probability of that pixel belonging to our object. 

How do we do it ? We create a histogram of an image containing our object of interest (in our case, the ground, leaving player and other things). The object should fill the image as far as possible for better results. And a color histogram is preferred over grayscale histogram, because color of the object is a better way to define the object than its grayscale intensity. We then “back-project” this histogram over our test image where we need to find the object, ie in other words, we calculate the probability of every pixel belonging to the ground and show it. The resulting output on proper thresholding gives us the ground alone.

It usually involves using a HSV image, as its color can be effectively specified as 2D, instead of 3D for RGB

```python
marbles = cv2.imread("public-images/marbles.jpg")[:,:,::-1]

x, y, r = 540, 757, 30#30
marbles_withroi = cv2.circle(marbles.copy(), (x, y), r, color = (0,255,0), thickness = 5)
plt.imshow(marbles_withroi)
```

```python
#roi_mask = cv2.circle(np.zeros(marbles.shape[:-1], dtype=marbles.dtype), 
#                      (x, y), r, color = 255, thickness = -1)
#roi = cv2.bitwise_and(marbles, marbles, mask = roi_mask)
roi = marbles[y-r:y+r, x-r:x+r]
plt.imshow(roi)
```

```python
## convert to hsv
roi_hsv = cv2.cvtColor(roi, cv2.COLOR_RGB2HSV)
marbles_hsv = cv2.cvtColor(marbles, cv2.COLOR_RGB2HSV)


roi_hist = cv2.calcHist([roi_hsv], channels = [0, 1], 
                        mask = None, histSize = [180, 256],
                        ranges = [0, 180, 0, 256])
## normalize histogram and apply backprojection
cv2.normalize(roi_hist, roi_hist, 0, 255, cv2.NORM_MINMAX)
dst = cv2.calcBackProject([marbles_hsv], [0, 1], 
                          hist = roi_hist, ranges = [0, 180, 0, 256], 
                          scale = 1)
## convolute with circular disc
disc = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
dst = cv2.filter2D(dst, -1, disc)

## threshold and binary and
ret, thr = cv2.threshold(dst, 50, 255, cv2.THRESH_BINARY, )
plt.imshow(thr, cmap = plt.cm.gray)
```

```python
result = cv2.bitwise_and(marbles,  marbles, mask = thr)
plt.imshow(result)
```

```python
## in range solution -- ALMOST THE SAME!!
lower_color = np.array([np.min(roi_hsv[:,:,i]) for i in [0, 1, 2]])
upper_color = np.array([np.max(roi_hsv[:,:,i]) for i in [0, 1, 2]])
marbles_thr = cv2.inRange(marbles_hsv, lower_color, upper_color, )
plt.imshow(cv2.bitwise_and(marbles, marbles, mask = marbles_thr))
```

```python
marbles = cv2.imread("public-images/marbles.jpg")[:,:,::-1]

x1, y1, r1 = 540, 757, 30#30
x2, y2, r2 = 1282, 729, 30
marbles_withroi = cv2.circle(marbles.copy(), (x1, y1), r1, color = (0,255,0), thickness = 5)
marbles_withroi = cv2.circle(marbles_withroi, (x2, y2), r2, color = (0,255,0), thickness = 5)
plt.imshow(marbles_withroi)
```

```python
roi1 = marbles[y1-r1:y1+r1, x1-r1:x1+r1]
roi2 = marbles[y2-r2:y2+r2, x2-r2:x2+r2]
roi = np.hstack([roi1, roi2])
plt.imshow(roi)
```

```python
## convert to hsv
roi_hsv = cv2.cvtColor(roi, cv2.COLOR_RGB2HSV)
marbles_hsv = cv2.cvtColor(marbles, cv2.COLOR_RGB2HSV)


roi_hist = cv2.calcHist([roi_hsv], channels = [0, 1], 
                        mask = None, histSize = [180, 256],
                        ranges = [0, 180, 0, 256])
## normalize histogram and apply backprojection
cv2.normalize(roi_hist, roi_hist, 0, 255, cv2.NORM_MINMAX)
dst = cv2.calcBackProject([marbles_hsv], [0, 1], 
                          hist = roi_hist, ranges = [0, 180, 0, 256], 
                          scale = 1)
## convolute with circular disc
disc = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
dst = cv2.filter2D(dst, -1, disc)

## threshold and binary and
ret, thr = cv2.threshold(dst, 50, 255, cv2.THRESH_BINARY, )
plt.imshow(thr, cmap = plt.cm.gray)

result = cv2.bitwise_and(marbles,  marbles, mask = thr)
plt.figure()
plt.imshow(result)
```
