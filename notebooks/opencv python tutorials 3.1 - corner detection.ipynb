## Harris Corner Detection
- corners are regions in the image with large variation in intensity in all the directions.
- it cannot be directly used as features
- `cv2.cornerHarris()`: cornerHarris(src, blockSize, ksize, k[, dst[, borderType]]) -> dst
- blockSize specifies neighborhood for change of pixel, ksize for calculation of derivative in sobel, k is free parameter

## Shi-Tomasi Good Feature to Track
- modification to Harris Corner, it picks N strongest corners
- goodFeaturesToTrack(image, maxCorners, qualityLevel, minDistance[, corners[, mask[, blockSize[, useHarrisDetector[, k]]]]]) -> corners

#### Harris Corner

```python
sudoku = cv2.imread("public-images/fish.png")[:,:,::-1]
gray_sudoku = cv2.cvtColor(sudoku, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_sudoku, cmap = plt.cm.gray)
```

```python
corners = cv2.cornerHarris(gray_sudoku, 2, 3, 0.0001)
corners = cv2.dilate(corners, np.ones((5, 5)))
corners = corners >= 0.0001*corners.max()
## dilate to make it more visible
```

```python
img = sudoku.copy()
img[corners] = (255, 0, 0)
plt.imshow(img)
```

```python
def plot_corner(k):
    corners = cv2.cornerHarris(gray_sudoku, 2, 3, k = 10**k)
    corners = cv2.dilate(corners, np.ones((5, 5)))
    corners = corners >= 0.0001*corners.max()
    plt.imshow(corners)
    
widgets.interact(plot_corner, k = (-5, 2, 0.1))
```

#### Good Features to Track

```python
corners = cv2.goodFeaturesToTrack(gray_sudoku, 15, 0.1, 100, )

img = sudoku.copy()

for c in corners[:,:,:]:
    x, y = map(int, c[0])
    img = cv2.circle(img, (x, y), 10, color = (0, 255, 0), thickness = -1)
    
plt.imshow(img)
```
