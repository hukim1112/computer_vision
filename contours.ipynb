## Finding Contours
- A contour is a curve of points, with no gaps in the curve. 
- Contours are extremely useful for such things as shape approximation and analysis.
- In order to find contours in an image, you need to first obtain a binarization of the image, using either ***edge detection***
methods or ***thresholding***.
- [doc for canny](http://docs.opencv.org/trunk/doc/py_tutorials/py_imgproc/py_canny/py_canny.html)

```python
import skimage 
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
## load image
coins = cv2.imread("data/coins.jpg")[:,:,::-1]
plt.imshow(coins)
```

```python
## gray scale
gray_coins = cv2.cvtColor(coins, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_coins, cmap = plt.cm.gray)
```

```python
## blurring by median
blur_coins = cv2.medianBlur(gray_coins, 11)
plt.imshow(blur_coins, cmap = plt.cm.gray)
```

```python
## get the edge - NOTICE the noise on the edges - what is what contours should remove
## you should always play with the threshold minValue and maxValue
edge_coins = cv2.Canny(blur_coins, 30, 30*3)
plt.imshow(edge_coins, cmap = plt.cm.gray)
```

```python
## find contours
img, contours, hierarchy = cv2.findContours(edge_coins, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
#print contours.shape
```

```python
plt.imshow(img, cmap = plt.cm.gray)
```

```python
plt.imshow(cv2.drawContours(gray_coins, contours, -2, (0, 255, 0), 2))
```

```python
## skimage version
from skimage import measure
contours = measure.find_contours(edge_coins, 200, )
```

```python
plt.imshow(coins)
for i, contour in enumerate(contours):
    plt.plot(contour[:, 1], contour[:, 0])
```
